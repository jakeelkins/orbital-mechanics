```python
'''
takes SMA, e, b as params, varies E
'''
a = 1.55e+8    #km
e = 0.975
b = a*np.sqrt(1- e**2)
```

```python
a*(1+e)
```

```python
E_list = np.linspace(0, 2*np.pi, 1000)

r_e_list = []
r_p_list = []

for E in E_list:
    
    r_e = a*(np.cos(E) - e)
    r_p = b*np.sin(E)
    
    r_e_list.append(r_e)
    r_p_list.append(r_p)
    
```

```python
a = 1986.18    #km
b = 8766.784
e = 4.533
p = 38826.02
```

```python
theta_list = np.linspace(-np.pi/2, np.pi/2, 2000)

r_e_list = []
r_p_list = []

for theta in theta_list:
    
    r = abs(p/(1+(e*np.cos(theta))))
    
    r_e = r*np.cos(theta)
    r_p = r*np.sin(theta)
    
    r_e_list.append(r_e)
    r_p_list.append(r_p)
    
```

```python
omega = 45*np.pi/180
theta = 260*np.pi/180
inclin = 35*np.pi/180
```

```python
DCM = np.array([
    [np.cos(omega)*np.cos(theta) - np.sin(omega)*np.cos(inclin)*np.sin(theta),
    -np.cos(omega)*np.sin(theta) - np.sin(omega)*np.cos(inclin)*np.cos(theta),
    np.sin(omega)*np.sin(inclin)],
    [np.sin(omega)*np.cos(theta) + np.cos(omega)*np.cos(inclin)*np.sin(theta),
    -np.sin(omega)*np.sin(theta) + np.cos(omega)*np.cos(inclin)*np.cos(theta),
    -np.cos(omega)*np.sin(inclin)],
    [np.sin(inclin)*np.sin(theta), np.sin(inclin)*np.cos(theta), np.cos(inclin)]
])
```
